#Define the features wil be used and the target(output)

```python

X = df[['Age', 'PhysicalActivity', 'SleepQuality', 'MemoryComplaints', 'FamilyHistoryAlzheimers', 'Depression', 'Diabetes', 'Hypertension', 'CholesterolTotal', 'MMSE']]
y=df['Forgetfulness']

X.head()
```

#Check if there is categorial columns

```python


object_cols=[col for col in X.columns if X[col].dtype=="object"]

if not object_cols:  # check if the list is empty
    print("No categorial columns")
else:
    print("categorial columns found")
```

# Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
train_X.head()
```

# Feature scaling

```python


from sklearn.preprocessing import MinMaxScaler,Normalizer,StandardScaler

scaler=MinMaxScaler(feature_range=(0,1))
train_X_scaled=scaler.fit_transform(train_X)
test_X_scaled=scaler.transform(test_X)
```

```python
print(pd.DataFrame(train_X_scaled,columns=train_X.columns))
```

#Finding the best K value

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score

f1_scores = []
values = range(1, 40, 2)

for K in values:
    model = KNeighborsClassifier(n_neighbors=K,metric="manhattan",weights="distance")
    model.fit(train_X_scaled, train_y)
    pred = model.predict(test_X_scaled)
    score = f1_score(test_y, pred, average='weighted')  #weihted for class imbalance
    f1_scores.append(score)
    print(f"F1-score for k = {K}: {score:.4f}")
```

# Plotting the K values

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(values, f1_scores, label='F1-Score', color='blue', marker='o', markerfacecolor='red', markersize=8)
plt.xlabel('K Value')
plt.ylabel('F1-Score')
plt.title('F1-Score vs K Value for KNN')
plt.grid()
plt.show()
```

# Appying KNN with the best K value

```python
from sklearn.metrics import accuracy_score,classification_report



knn_model =KNeighborsClassifier(n_neighbors = 13,metric="manhattan",weights="distance")
knn_model.fit(train_X_scaled,train_y)

knn_pred=knn_model.predict(test_X_scaled)
knn_acc=accuracy_score(knn_pred,test_y)
knn_acc_percent = knn_acc * 100
print(f"KNN Accuracy: {knn_acc_percent:.2f}%")


print(classification_report(test_y, knn_pred))
```

# Calculating best n_estimator for Random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

val = []
val.clear()
def best_n_estimator(n_estimator):
    model = RandomForestClassifier(n_estimators=n_estimator,random_state=42, class_weight='balanced')
    model.fit(train_X_scaled, train_y)
    pred = model.predict(test_X_scaled)
    acc = accuracy_score(test_y, pred)
    acc = acc * 100
    val.append(acc)
    return acc  # Return accuracy for direct use

x = [10, 20, 30, 40, 50, 100, 200, 300, 400,500]

for i in x:
    acc = best_n_estimator(i)
    print(f"Estimator: {i}, accuracy is: {acc:.2f}%")


```

# Plotting the n estimators

```python
plt.figure(figsize=(10,6))
plt.plot(x,val, label='val', color='black', marker='o', markerfacecolor='green', markersize=8)
plt.xlabel('Estimators')
plt.ylabel('Accurancy')
plt.title('Estimators and their accurancy ')

plt.grid()
plt.show()
```

# Applying RandomForest with the best n estimator


```python
randomforest_model=RandomForestClassifier(n_estimators=100,random_state=42, class_weight='balanced')
randomforest_model.fit(train_X_scaled,train_y)
randomforest_pred=randomforest_model.predict(test_X_scaled)
random_forest_acc=accuracy_score(randomforest_pred,test_y)
random_forest_acc_percent=random_forest_acc*100

results_df = pd.DataFrame({
    'Actual': test_y,      # Actual values
    'Predicted': randomforest_pred  # Predicted values
})

# Display the results
print(results_df.head(5))


```

# Comparing KNN and RandomForest

```python
models = ['Random Forest', 'KNN ']
accuracies = [random_forest_acc_percent, knn_acc_percent]


plt.figure(figsize=(8, 5))
plt.bar(models, accuracies, color=['green', 'blue'])
plt.ylim(0, 100)
plt.title('Model Accuracy Comparison', fontsize=14)
plt.xlabel('Models', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
